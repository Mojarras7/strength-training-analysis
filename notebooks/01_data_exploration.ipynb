```python
import pandas as pd
import numpy as np

def get_raw_preview(file_path):
    """
    Loads a raw CSV without headers to inspect its structure.
    """
    return pd.read_csv(file_path, header=None)

    # Paths to your files
file_b5 = '../data/raw/Progreso - Bloque 5.csv'
file_b8 = '../data/raw/Progreso - Bloque 8.csv'


# Load previews
df5_raw = get_raw_preview(file_b5)
df8_raw = get_raw_preview(file_b8)

# Find the anchor row for 'EJERCICIOS' in both
# We search in the 3rd column (index 2) where exercise names usually live
row_b5 = df5_raw[df5_raw.iloc[:, 2] == 'EJERCICIOS'].index[0]
row_b8 = df8_raw[df8_raw.iloc[:, 2] == 'EJERCICIOS'].index[0]

print(f"Block 5 header starts at row: {row_b5}")
print(f"Block 8 header starts at row: {row_b8}")

```
